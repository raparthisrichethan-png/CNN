```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_path="C:/Users/Sri Chethan/OneDrive/Desktop/ML/CNN/PetImages"
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,

    validation_split=0.2
)
```

```python
train_data = datagen.flow_from_directory(
    dataset_path,target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='training'
    )
```

```python
val_data=datagen.flow_from_directory(
    dataset_path,target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
    )
```

```python
print(train_data.class_indices)
```

```python
model = keras.Sequential()
```

```python
# first convolution layer
model.add(
    keras.layers.Conv2D(
        filters=32,
        kernel_size=(3,3),
        activation='relu',
        input_shape = (128,128,3)
    )
)
model.add(keras.layers.BatchNormalization())
```

```python
# first pooling layer
model.add(
    keras.layers.MaxPooling2D(pool_size=(2,2))
)
```

```python
# 2 convolution layer
model.add(
    keras.layers.Conv2D(
        filters=64,
        kernel_size=(3,3),
        activation='relu',
    )
)
model.add(keras.layers.BatchNormalization())
```

```python
# second pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
model.add(
    keras.layers.Conv2D(
        filters=128,
        kernel_size=(3,3),
        activation='relu',
    )
)
model.add(keras.layers.BatchNormalization())
```

```python
#  pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# 4th convolution layer
model.add(
    keras.layers.Conv2D(
        filters=128,
        kernel_size=(3,3),
        activation='relu',
    )
)
model.add(keras.layers.BatchNormalization())
```

```python
# 4th pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# 5th convolution layer
model.add(
    keras.layers.Conv2D(
        filters=256,
        kernel_size=(3,3),
        activation='relu',
    )
)
model.add(keras.layers.BatchNormalization())
```

```python
# 5th pooling layer
model.add(
    keras.layers.MaxPooling2D(
        pool_size=(2,2)
    )
)
```

```python
# Dropout for regularization
model.add(
    keras.layers.Dropout(0.5)
)
```

```python
model.add(
    keras.layers.Flatten()
)
model.add(
    keras.layers.Dense(
        units=256,
        activation='relu'
    )
)
model.add(keras.layers.Dropout(0.5))
model.add(
    keras.layers.Dense(
        units=128,
        activation='relu'
    )
)
model.add(keras.layers.Dropout(0.3))
```

```python
model.add(
    keras.layers.Dense(
        units=1,
        activation='sigmoid'
    )
)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_data,
    epochs=5,
    validation_data=val_data
)

loss,accuracy=model.evaluate(val_data)
print("Accuracy :",accuracy)
```
